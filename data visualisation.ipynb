# Multiple Lines Plotting

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x=np.arange(1,12)
y1=x+2
y2=x+6
y3=x+5
plt.plot(x,y1,color='black')
plt.plot(x,y2,color='red')
plt.plot(x,y3,color='yellow')
plt.grid(True)
plt.xlabel("this is x label")
plt.ylabel("this is y label")
plt.title("multiple lines plotting")
plt.show()
```

# Bar Chart Plotting

```python
import matplotlib.pyplot as plt
plt.bar([1,2,3],[5,6,7])
plt.xlabel("this is x-axis label")
plt.ylabel("this is y-axis label")
plt.title("simple bar chart")
plt.show()
```

# Case study


```python
student_name=['john','sohil','monika','peter']
index=np.arange(len(student_name))
marks=[75,89,60,79]
plt.bar(index,marks)
plt.xlabel("student_name")
plt.ylabel("marks")
plt.title("students performance")
plt.xticks("index,student_name")
plt.show()
```

# Pie Chart Plotting

```python
percentage=[50,30,20]
language=['python','java','c++']
plt.pie(percentage,labels=language)
plt.title('simple pie chart')
plt.show()
```
